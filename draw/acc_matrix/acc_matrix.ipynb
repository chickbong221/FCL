```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the 5x5 matrix with given values
data = np.array([
    [0.2947, 0.18245, 0.1883, 0.18695, 0.2389],
    [0.2071, 0.29695, 0.2187, 0.23245, 0.2094],
    [0.2193, 0.2255, 0.2569, 0.21785, 0.2053],
    [0.21425, 0.2253, 0.21635, 0.304, 0.22],
    [0.25185, 0.21845, 0.2186, 0.2564, 0.28695]
])

# Create a heatmap with values formatted to 4 decimal places
plt.figure(figsize=(6, 5))
sns.heatmap(data, annot=True, fmt=".4f", cmap="Blues", cbar=False,
            linewidths=0.5, linecolor='black', square=True,
            xticklabels=range(5), yticklabels=range(5))

# Set axis labels
plt.xlabel("Prediction")
plt.ylabel("Task")

# Improve layout
plt.title("Heatmap of AFFCL")
plt.tight_layout()
plt.show()
```
